```python
import pandas as pd
import os
import kagglehub

path = kagglehub.dataset_download("christopheiv/winemagdata130k")
cv_path = os.path.join(path,"winemag-data-130k-v2.csv")

reviews = pd.read_csv(cv_path, index_col=0)
pd.set_option('display.max_rows', 5) # Forma de visualizar la salida
```

# Groupwise analysis

Replicando value_counts() usando groupby y count

```python
reviews.groupby('points').points.count()
# Explicacion
# EL primer 'points', indica al groupby que debe agrupar po points
# El segundo points, se esta eligiendo la columna sobre el cual se aplicara la funcion count
```

```python
reviews.head()
```

```python
reviews.groupby('country').winery.count()
```

Al aplicar una nueva operacion con el groupby -> Se genera una nueva serie o un nuevo DF 
A el DF o Serie se le puede aplicar directamente apply() y podemos manipular la data 
*Ejemplo* Obtener el nombre del primer vino con resenhado en cada winery

```python
reviews.groupby('winery').apply(lambda df: df.title.iloc[0])
```

```python
reviews[['winery', 'title']].sort_values(by='winery')
```

```python
Se puede agrupar por mas de una columna
```

Ejemplo: Mejor vino del pais y provincia

```python
reviews.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()])
```

.agg() para poder agregar diferentes funciones en un DF

```python
#Ejemplo: Generar un pequenho summary
reviews.groupby(['country']).price.agg([len, min, max])
```

# Multi-indexes
Dependiendo de la funcion que ejecutemos, a veces generara multiindices, que se diferencia del index en que posee multiples niveles

```python
countries_reviewed = reviews.groupby(['country', 'province']).description.agg([len])
countries_reviewed
```

```python
mi = countries_reviewed.index
type(mi)
```

Multiindex posee varios metodos para gestionar su estructura, sin embargo el mas usado es reset_index() que lo convierte a un indice normal

```python
countries_reviewed.reset_index()
```

# Sorting
sort_values()

```python
countries_reviewed = countries_reviewed.reset_index()
countries_reviewed.sort_values(by='len')
```

```python
countries_reviewed.sort_values(by='len', ascending=False)
```
