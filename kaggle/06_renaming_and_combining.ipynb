# Renaming

Funciones para canviar nombres 
* .rename(): Puede cambiar nombre de columnas o indices
* .rename_axis

```python
reviews.rename(columns={'points':'score'})
```

```python
reviews.rename(index={0: 'firstEntry', 1:'secondEntry'})
```

Para cambiar index, es mas conveniente usar set_index()

```python
reviews.head()
```

```python
reviews.rename_axis("wines", axis='rows')
```

```python
reviews.rename_axis("wines", axis='rows').rename_axis("fields", axis='columns')
```
